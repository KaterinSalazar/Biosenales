2.   Los dos vectores deben tener la misma longitud







```python
if isinstance(a, np.ndarray) and isinstance(b, np.ndarray) and len(a) == len(b): #comprueba que los dos arrays sean un objeto np y tengan la misma longitud
  resultado = np.dot(a, b)
  print(resultado)
else:
  print("a o b no es un np.array, o no tienen la misma longitud")
```

### **1) Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.**

```python
datos = {
    "Ensayo 1": 3.5,
    "Ensayo 2": 4.0,
    "Ensayo 3": 2.8,
    "Ensayo 4": 4.5
}
def notas(dic):
  notas=pd.Series(dic)
  estadisticas = {
    "m√≠nimo": notas.min(),
    "m√°ximo": notas.max(),
    "media": notas.mean(),
    "desviaci√≥n tipica": notas.std()
}
  return pd.Series(estadisticas)
notas(datos)
```
