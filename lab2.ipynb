```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def impseq(n0, n1, n2):

# Genera x(n) = delta(n-n0); n1 <= n <= n2

# ----------------------------------------------

  n = np.arange(n1,n2+1) # Se crea el vector de muestras

  x = (n-n0) == 0

  return [x,n]

print(impseq(3,-3,3))
```

### **FUNCION ESCALON UNITARIO**

```python
def unitario(n0,n1,n2):
  n=np.arange(n1,n2+1)
  x=np.where(n>=n0,1,0)

  return [x,n]
```

### **FUNCION RAMPA**

```python
def rampa(m,n0,n1,n2):
  n=np.arange(n1,n2+1)
  x=np.where(n<n0,0,m*(n-n0))

  return [x,n]
```

### **SECUENCIAS USANDO FUNCIONES DE PYTHON**

a. el vector de muestras debe abarcar todas las muestras tomadas en la operacion, en este caso en el intervalo [-3,0]

```python
y1,n=impseq(-1,-3,0)
y2,_=impseq(-3,-3,0)
y3,_=impseq(-2,-3,0)
y4,_=impseq(0,-3,0)
x1=3*y1+5*y2+3*y3+4*y4
plt.stem(n, x1)
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.title("Impulso discreto")
plt.grid()
plt.show()
```

```python
u1,n=unitario(-4,-6,6)
u2,_=unitario(-2,-6,6)
x2=-2*u1 +3*u2
plt.stem(n, x2, linefmt='r-', markerfmt='ro', basefmt='k-')
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.title("ESCALONES UNITARIOS")
plt.grid()
plt.show()
```

```python
n=np.arange(0,101)
x1=np.exp(0.01*n)
x2=np.sin(2*np.pi*n)
x3=x1*x2
plt.stem(n, x3, linefmt='m-', markerfmt='m', basefmt='m-')
plt.xlabel("n")
plt.ylabel("x3[n]")
plt.title("multiplicacion de funciones")
plt.grid()
plt.show()
```

```python
r1,n=rampa(1,-3,-10,10)
r2,_=rampa(1,2,-10,10)
u1,_=unitario(3,-10,10)
x4=2*r1-r2-5*u1
plt.stem(n, x4)
plt.xlabel("n")
plt.ylabel("x4[n]")
plt.grid()
plt.show()
```

## **4. DERIVADA**
la derivada se define como la tasa de cambio en una funcion, pero en funciones discretas se tienen puntos, por lo nque la uniza manera de medir esta tasa es restando al valor de la posicion n+1 el valor n

```python
r1,n=rampa(1,0,0,17)
r2,_=rampa(1,5,0,17)
r3,_=rampa(1,10,0,17)
x=np.where((n>=0) & (n<=5),r1,np.where((n>=6 ) & (n<=11),r2,r3))

#DERIVADA DISCRETA
y = np.diff(x, prepend=0)

plt.figure(figsize=(10,6))

# -------- Seal original --------
plt.subplot(2,1,1)
plt.stem(n, x, linefmt='b-', markerfmt='bo', basefmt='k-')
plt.title("Seal Original x[n]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()

# -------- Derivada discreta --------
plt.subplot(2,1,2)
plt.stem(n, y, linefmt='r-', markerfmt='ro', basefmt='k-')
plt.title("Derivada Discreta y[n]")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()

plt.tight_layout()
plt.show()
```
